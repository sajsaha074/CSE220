```python
class Patient:
    def __init__(self, id, name, age, blood_group):
        self.id = id
        self.name = name
        self.age = age
        self.blood_group = blood_group
        self.next = None
        self.prev = None

class WRM:
    def __init__(self):
        self.head = None
    def register_patient(self, id, name, age, blood_group):
        new = Patient(id, name, age, blood_group)
        if self.head is None:
            self.head =new
            self.head.next = self.head
            self.head.prev = self.head
        else:
            last = self.head.prev
            last.next = new
            new.prev = last
            new.next = self.head
            self.head.prev = new

    def serve_patient(self):
        if self.head is None:
            print("No patients in the waiting room.")
        else:
            served = self.head
            if self.head.next == self.head:  # Only one patient in the waiting room
                self.head = None
            else:
                self.head = self.head.next
                last = self.head.prev
                last.next = self.head
                self.head.prev = last
            print("Serving patient:", served.name)

    def cancel_all(self):
        if self.head is None:
            print("No appointments to cancel.")
        else:
            self.head = None
            print("All appointments cancelled.")

    def can_doctor_go_home(self):
        if self.head is None:
            return True
        else:
            return False

    def show_all_patients(self):
        if self.head is None:
            print("No patients in the waiting room.")
        else:
            current = self.head
            while True:
                print("Patient ID:", current.id)
                current = current.next
                if current == self.head:
                    break

#Driver code
Wrm=WRM()
while True:
    print("\nOptions:")
    print("1. Add Patient")
    print("2. Serve Patient")
    print("3. Show All Patients")
    print("4. Can Doctor Go Home?")
    print("5. Cancel All Appointments")
    print("6. Exit")
    option = int(input("Enter your choice: "))
    if option == 1:
        id = int(input("Enter patient ID: "))
        name = input("Enter patient name: ")
        age = int(input("Enter patient age: "))
        blood_group = input("Enter patient blood group: ")
        Wrm.register_patient(id, name, age, blood_group)
        print("Patient added successfully.")
    elif option == 2:
        Wrm.serve_patient()
    elif option == 3:
        Wrm.show_all_patients()
    elif option == 4:
        if Wrm.can_doctor_go_home():
            print("Yes, can go home.")
        else:
            print("No, there are patients.")
    elif option == 5:
        Wrm.cancel_all()
    elif option == 6:
        break
    else:
        print("Invalid choice")
```
